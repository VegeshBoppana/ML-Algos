# Rain Prediction — Tree-Based & Ensemble Methods

## Background

In previous Assingment , we tackled the binary classification problem of predicting whether  it will rain tomorrow (RainTomorrow: Yes/No) using the Australian Weather dataset (weatherAUS.csv), which contains 145k daily weather observations collected across 49 locations in Australia between 2007 and 2017.
The dataset presented several challenges including severe class imbalance (78% No Rain, 22% Rain), missing values across multiple features, and highly correlated meteorological measurements.


After careful preprocessing — dropping missing target rows, applying grouped imputation strategies (Median for high-missing columns, KNN Imputer for low-missing numerical columns, Mode for categorical columns), cyclical encoding for wind directions and month, target encoding for Location, binary encoding for RainToday/RainTomorrow, and a time-based train/test split (Train: 2007–2014, Test: 2015–2017) — we benchmarked Logistic Regression, KNN, Naive Bayes, and SVM. Class imbalance was handled using SMOTE+ENN combined with threshold tuning. The best model was SVM Linear + Threshold Tuning, achieving F1=0.63, Recall=0.70, and AUC-ROC=0.86 on the test set.


In this assignment, we extend the same problem using tree-based and ensemble methods — Decision Tree Classifier, Random Forest Classifier, and XGBoost — with Optuna-driven hyperparameter tuning using Stratified 5-fold cross-validation, and model interpretability via SHAP analysis.


The preprocessing pipeline is adapted to suit
tree-based models: no feature scaling (trees are invariant to feature magnitude), no Yeo-Johnson transformation (trees are invariant to monotonic transforms), and all features retained including correlated ones (trees handle multicollinearity natively through split selection). Class imbalance is handled primarily via class_weight='balanced', which directly integrates into the split-finding process of tree models — a more principled approach than synthetic resampling for this model family. All models are
evaluated using F1-Score as the primary metric, **with Recall prioritized given that missing a rain event carries significantly higher real-world cost than a false alarm.**
